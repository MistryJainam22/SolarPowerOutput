 # Let's Build Machine Learning Model To Predict Solar Power Using Linear  Regression

```python
# Linear Regresssion Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,median_absolute_error
```

```python
df.columns
```

```python
#splitting the dataset into fratures and target
x = df. drop('generated_power_kw', axis=1)
y = df['generated_power_kw']
```

```python
# Let,s split he dataset into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
# Scaling
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Train Model
model = LinearRegression()
model.fit(x_train_scaled, y_train)
```

```python
#Evaluated the model - test
y_pred = model.predict(x_test_scaled)
mae = median_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error on Test Set:  {mae}")
```

```python
# Evaluate the model on the training set
y_pred_train = model.predict(x_train_scaled)
mae_train = median_absolute_error(y_train, y_pred_train)
print(f"Mean Absolute Error on Test set: {mae_train}")
```
